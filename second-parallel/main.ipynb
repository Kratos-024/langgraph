```python
class EvalutionSchema(BaseModel):
    feedback:str=Field(description='Detailed feedback for the essay')
    score:str=Field(description='Score out of 10',g=0,l=10)
```

```python
import operator


class UPSCState(TypedDict):
    eassy:str
    lang_feed:str
    anal_feed:str
    cot:str
    overall_feedback:str
    individual_scores:Annotated[list[int],operator.add]
    avg_score:float
```

```python
def   evalute_langauge(state:UPSCState):
    promp="Evalute the language quality of the following eassy and provide a feedback and assign a score out of 10\n {eassy}"
    output = llm.invoke(promp)
    return {'language',output.feedback,"individual_score":output.score}
```

```python
graph = StateGraph(UPSCState)
graph.add_node('evalute_langauge',evalute_langauge)
```
